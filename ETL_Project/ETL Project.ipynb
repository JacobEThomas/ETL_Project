```python
#Import Dependencies Read Existing CSV and Define Variables
import pandas as pd
import requests as req
import time as time
df = pd.read_csv('Data1_CSV.csv')
api_key = "2b82346ed86d1f148705d37ae2c14ee7"
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 
lat = []
lon = []
counter = 0
df.head()
```

```python
#Count Unique Zip Codes
df['zip_8'].nunique()
```

```python
#Create New DF With Target Variables
df2 = pd.DataFrame(columns =[ "zip_8", "Latitude","Longitude"])
```

```python
#Import Unique Zip Codes From Existing Data
df2["zip_8"]=df.zip_8.unique()
```

```python
#Check for Proper Importation
df2.head()
```

```python
#Loop Through Zip Codes to Identify Lattitude and Longitude from OpenWeatherAPI
for index, row in df2.iterrows():
    counter = counter + 1
    zip = row["zip_8"]
    target_url = url + "&q=" + zip
    print("Looping through Zip Code # "+str(counter))
    print("Zip Code is "+row["zip_8"])
    try:
        info = req.get(target_url).json()
        lon.append(info['coord']['lon'])
        lat.append(info['coord']['lat'])
        #Set sleep timer to allow for complete calls - API limits to 60 per minute
        time.sleep(1.01)
    except:
        lon.append("NaN")
        lat.append("NaN")
        continue
        
print("____________________________________________")
print("Loop Complete")
print("____________________________________________")
```

```python
#Transform values from arrays to DF
df2["Latitude"]=lat
df2["Longitude"]=lon
df2.head()
```

```python
#Merge Existing Data with Newly Generated API Lat/Long Data
df3 = pd.merge(df, df2, on='zip_8')
df3.head()
```
